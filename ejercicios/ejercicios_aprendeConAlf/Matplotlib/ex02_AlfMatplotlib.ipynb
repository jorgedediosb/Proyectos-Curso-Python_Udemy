02. Escribir una función que reciba un diccionario con las notas de las asignaturas de un curso y una cadena con el nombre de un color y devuelva un diagrama de barras de las notas en el color dado.

```python
import matplotlib.pyplot as mpl
```

```python
def notas_color(notas, color):
    fig, ax = mpl.subplots()
    ax.bar(notas.keys(), notas.values(), color = color)
    return ax
```

```python
notas = {'Programación':6, 'Matemáticas':5, 'Álgebra': 4, 'Historia':7}
```

```python
notas_color(notas, 'orange')
mpl.show()
```
