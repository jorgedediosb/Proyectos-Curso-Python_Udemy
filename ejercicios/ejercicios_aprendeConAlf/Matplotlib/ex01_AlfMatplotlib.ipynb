01. Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla un diagrama de líneas con la evolución de las ventas.

```python
import matplotlib.pyplot as mpl
```

```python
# %matplotlib inline hace que se visualice el diagrama sin la línea del objeto:
%matplotlib inline
```

```python
inicio = int(input('Introduce el año de inicio: '))
```

```python
fin = int(input('Introcude el año final: '))
```

```python
ventas = {}
```

```python
for i in range(inicio, fin+1):
    ventas[i] = float(input('Ventas año ' + str(i) + ':'))
```

```python
fig, ax = mpl.subplots()
ax.plot(ventas.keys(), ventas.values())
mpl.show()
```
